```python
# number of octaves = 4, number of scale levels = 5, initial σ=1.6, k=2–√ 
#sigma = 1.6
#octaves = 4
#采用一个octave中取样3张图片的形式（在最后的DoG中找极值点的时候用中间的三张），但是必须要额外生成3张来保证在计算过程中尺度的连续性
#scales = 6#最后在DoG的中间两幅图像中找极值点
#k = math.pow(2, 1 / (scales - 3))

def __sigma(sigma, octaves, scales, k):
    sig = []
    for o in range(octaves):
        sig.append([sigma*math.pow(2,o)*math.pow(k,s) for s in range(scales)])
    sig = np.array(sig, copy=False)
    
    return sig

#sig = __sigma(sigma, octaves, scales, k)
#sig
```

```python
def __gaussianPyramid(base, sig, octaves, scales, kern_size=(3, 3)):
    #pyramid 是五维数组
    pyramid = np.zeros((octaves, scales), dtype=type(base))
    pyramid[0, 0] = base
    for o in range(octaves):
        if o != 0:
            #下一个octave的第一个sigma是上一个octave的sigma的两倍
            #下一个octave的newBase是上一个octave中使用相同sigma的那一张图长宽缩小为二分之一之后的图像
            #上一个octave中使用相同sigma的一张图刚好是上一个octave中第scales-3个，因为每个octave多
            #生成3张模糊过的图片，具体自己推导出来就知道
            newBase = __zoom(pyramid[o-1, scales-3], 2, 2)
            pyramid[o, 0] = newBase
        #每个octave的第一张图片在进入这个循环之前已经确定了
        for s in range(scales):
            #因为是把pyramid的一行取出来进行添加，这一行取出来之后的第0个轴相当于没去出来的时候pyramid
            #的第1个轴
            blur = __gaussianBlur(pyramid[o, 0], sig[o, s], kern_size)
            pyramid[o, s] = blur
    return pyramid
            
def __zoom(oldBase, xFactor, yFactor):
    #暂时只接受缩小两倍
    #Factor 代表srcWidth/destWidth或者srcHeight/destHeight
    #暂时只使用双线性插值法来进行灰度图缩放
    oldX = oldBase.shape[0]
    oldY = oldBase.shape[1]
    newX = int(oldBase.shape[0]/xFactor)
    newY = int(oldBase.shape[1]/yFactor)
    newBase = np.zeros((newX, newY))
    for dstX in range(newX):
        for dstY in range(newY):
            srcX = dstX*xFactor
            srcY = dstY*yFactor
            newBase[dstX, dstY] = oldBase[srcX, srcY]
    return newBase

#不能in-place进行
#一种方式是使用二维核，另一种方式是使用x核y方向的两个线性核，先用x方向的进行卷积，之后对结果用y方向的进行卷积
#还有边缘的像素填充
def __gaussianBlur(base, sigma, kern_size=(3, 3)):
    assert kern_size[0] > 0 and kern_size[0] % 2 == 1 and kern_size[1] > 0 and kern_size[1] % 2 == 1, \
        "Kern Size must be positiv and odd"
    kern = __kernNormalization(__gaussianKern(sigma, kern_size))
    blur = np.zeros((base.shape[0], base.shape[1]))
    #用卷积核进行滑窗计算的时候，边缘的一部分像素暂时不处理，因为卷积核会超出图像范围
    rangeX = kern_size[0] // 2
    rangeY = kern_size[1] // 2
    #base的列数是x轴，行数是y轴，图像长宽都是362个像素，像素的Index最大是361，最小是0
    #base.shape[1]-rangeX不用加1
    #blur边缘没有处理的点保留为0
    for i in range(rangeX, base.shape[0]-rangeX):
        for j in range(rangeY, base.shape[1]-rangeY):
            blur[i, j] = np.sum(base[i-rangeX:i+rangeX+1, j-rangeY:j+rangeY+1] * kern)
    return blur
    
def __gaussianKern(sigma, kern_size):
    rangeX = kern_size[0]//2
    rangeY = kern_size[1]//2
    sig_kons = 1/(2*sigma*sigma)
    pi_kons = sig_kons/np.pi
    return np.array([[pi_kons*np.exp((-x*x-y*y)*sig_kons) for y in range(-rangeY, rangeY+1)] for x in range(-rangeX, rangeX+1)])

def __kernNormalization(kern):
    #归一化的整体操作是先计算出来kern中所有元素的和，之后每一个元素除以这个和
    #卷积过程是这个像素及周围的像素与卷积核对位相乘，结果之和为这个像素的新值
    #归一化是为了保证计算出来的新值仍然是在0-255区间内的，而不是属于更大或者
    #更小的区间
    sum = np.sum(kern)
    for i in range(kern.shape[0]):
        for j in range(kern.shape[1]):
            kern[i, j] = kern[i, j] / sum
    return kern
```

```python
def __DoGPyramid(gaussianPyramid):
    octaves = gaussianPyramid.shape[0]
    #一副DoG由gaussianPyramid中两幅相邻图像相减而来，所以DoG的scales比gaussianPyramid的小1
    scales = gaussianPyramid.shape[1] - 1
    DoGPyramid = np.zeros((octaves, scales),dtype=type(gaussianPyramid[0, 0]))
    for o in range(octaves):
        for s in range(scales):
            #src2减src1，即相邻两张图像中sigma大的图像减sigma小的图像
            #之后用np.abs防止负数出现
            DoGPyramid[o, s] = np.abs(gaussianPyramid[o, s+1] - gaussianPyramid[o, s])
            #DoGPyramid[o, s] = np.abs(gaussianPyramid[o, s+1])
    return DoGPyramid
```

```python
class Extrema():
    def __init__(self, x, y, sigma, interval, octave, offset, value, size, orientation):
        self.pos =(x, y)
        self.img_sigma = sigma
        self.interval = interval
        self.octave = octave
        self.offset = offset
        self.value = value
        self.size = size
        self.orientation = orientation

def  __localExtremaDetection(DoG, sig, contrast_threshold, eigenvalueRatio=10):
    threshold = np.floor(0.5 * contrast_threshold / (DoG.shape[1] - 2) * 255) 
    extremaList  = []
    for i in range(DoG.shape[0]):
        #一个octave中从第二张图片开始寻找直到倒数第二张
        for j in range(1, DoG.shape[1]-1):
            #每次要对比3x3x3区域内，当前点是不是最值，所以x和y从1开始，到对应图的倒数第二个像素为止
            for x in range(1, DoG[i, j].shape[0]):
                for y in range(1, DoG[i, j].shape[1]):
                    tempExtrema = DoG[i, j][x, y]
                    #tempExtrema必须是它的26个邻域之中的最大值，并且大于阈值，为了减小噪音
                    if DoG[i, j][x, y] > threshold and \
                       np.argmax(DoG[i, j][x-1:x+2, y-1:y+2])  == 5:
                       #np.max(DoG[i, j-1][x-1:x+2, y-1:y+2]) < tempExtrema and \
                       #np.max(DoG[i, j+1][x-1:x+2, y-1:y+2]) < tempExtrema:
                        extrema = __extremaFilter(DoG, sig, i, j, x, y, contrast_threshold, eigenvalueRatio)
                        if extrema:
                            extremaList.append(extrema)
    return extremaList

def __extremaFilter(DoG, sig, i, j, x, y, contrast_threshold, eigenvalueRatio):
    offset = None
    #只有模块（module），类（class）和函数（def、lambda）才会引入新的作用域，
    #if/elif/else/、try/except、for/while 等语句则不会引入新的作用域，
    #即外部可以访问在这些语句内定义的变量。
    #5是默认的迭代参数，之后改成可以修改迭代参数的类型
    numOfScales = DoG.shape[1] - 2;
    imgShape = DoG[i, j].shape
    scales = DoG.shape[1] - 2
    for a in range(5):
        gradient = __getGradient(DoG, i, j, x, y)
        hessian = __getHessian(DoG, i, j, x, y)
        offset = -np.linalg.lstsq(hessian, gradient, rcond=None)[0]
        if abs(offset[0]) < 0.5 and abs(offset[1]) < 0.5 and abs(offset[2]) < 0.5:
            break
        x += int(round(offset[0]))
        y += int(round(offset[1]))
        j += int(round(offset[2]))
        #判断这个极值点在移动之后是否超出边界，如果超出应该舍去
        if x < 1 or x >= imgShape[0] - 1 or y < 1 or y >= imgShape[1] - 1 or \
           j < 1 or j > scales:
            return None
        if a >= 4:
            return None
    extremaValue = DoG[i, j][x, y] + 0.5 * np.dot(gradient, offset)
    if abs(extremaValue) * scales >= contrast_threshold:
        if not __isEdgeResponse(hessian, eigenvalueRatio):
            #keypoint的size要是它当前的sigma乘offset中在尺度方向偏离的offset所造成的sigma
            #原来的计算是从最原始的sigma=1.6，根据sigma = sigma_orignal ** (s/S)
            #小s是极值点尺度方向的坐标，是原来的图片的尺度加上offset的尺度 
            #extremun_update相当于这里的offset
            #size = sigma * (2 ** ((image_index + extremum_update[2]) / float32(num_intervals)))* (2 ** (octave_index + 1))
            #最后乘的octave_index相当于i，可能是考虑到降采样中的图片大小问题，但是在之后的函数里又除掉了，所以没写，如果写了
            #在生成描述符的时候要把size还原，注意这里是2的i加1次方，比下面的指数多1
            #这里由于已经知道图片对应的尺度，所以改写一下
            size = sig[i, j] * (2 ** (offset[2] / numOfScales));
            #极值点的坐标也要考虑降采样中，图片的尺寸缩小的问题，所以要相应的扩大
            #i相当于x， j相当于y（为啥这里和通常的写法相比是反过来的，这老哥不会写反了吧）
            #((j + extremum_update[0]) * (2 ** octave_index), (i + extremum_update[1]) * (2 ** octave_index))
            x_original = (x + offset[0]) * (2 ** i)
            y_original = (y + offset[1]) * (2 ** i)
            return Extrema(y_original, x_original, sig[i, j], j, i, offset, abs(extremaValue), size, None)
    return None    
        
def __getGradient(DoG, i, j, x, y):
    dx = 0.5 * (DoG[i, j][x + 1, y] - DoG[i, j][x - 1, j])
    dy = 0.5 * (DoG[i, j][x, y + 1] - DoG[i, j][x, j - 1])
    ds = 0.5 * (DoG[i, j + 1][x, y] - DoG[i, j - 1][x, j])
    return np.array([dx, dy, ds])

def __getHessian(DoG, i, j, x, y):
    dxx = DoG[i, j][x+1, y]+DoG[i, j][x-1, y]-2*DoG[i, j][x, y]
    dyy = DoG[i, j][x, y+1]+DoG[i, j][x, y-1]-2*DoG[i, j][x, y]
    dss = DoG[i, j+1][x, y]+DoG[i, j-1][x, y]-2*DoG[i, j][x, y]
    dxy = 0.25 * ((DoG[i, j][x+1, y+1]+DoG[i, j][x-1, y-1])-(DoG[i, j][x+1, y-1]+DoG[i, j][x-1, y+1]))
    dxs = 0.25 * ((DoG[i, j+1][x+1, y]+DoG[i, j-1][x-1, y])-(DoG[i, j-1][x+1, y]+DoG[i, j+1][x-1, y]))
    dys = 0.25 * ((DoG[i, j+1][x, y+1]+DoG[i, j-1][x, y-1])-(DoG[i, j-1][x, y+1]+DoG[i, j+1][x, y-1]))
    return np.array([[dxx, dxy, dxs],
                     [dxy, dyy, dys],
                     [dxs, dys, dss]])

def __isEdgeResponse(hessian, eigenvalueRatio):
    xyHessian = hessian[:2, :2]
    trace = np.trace(xyHessian)
    det = np.linalg.det(xyHessian)
    r = eigenvalueRatio
    if det > 0 and (trace**2)*r < ((r+1)**2)*det:
        return False
    return True

def show_keypoints(image, extremaList):
    print(len(extremaList))
    plt.figure('image')
    plt.imshow(image, cmap="gray")
    for extrema in extremaList:
        plt.plot(extrema.pos[0], extrema.pos[1], 'r+')
    plt.show()
```

```python
def __computeKeypointsWithOrientations(keypoints, gaussian_image, radius_factor=3, num_bins=36, peak_ratio=0.8, scale_factor=1.5):
    keypoints_with_orientation = []
    for keypoint in keypoints:
        octave = keypoint.octave
        interval = keypoint.interval
        imgShape = gaussian_image[octave, interval].shape
        img = gaussian_image[octave, interval]
        #先通过邻域内像素点的方向直方图决定大概的方向
        scale = scale_factor * keypoint.size
        radius = int(round(radius_factor * scale))
        weight_factor = -0.5 / (scale ** 2)
        raw_histogram = np.zeros(num_bins)
        smooth_histogram = np.zeros(num_bins)
        
        for i in range(-radius, radius + 1):
            y = int(round(keypoint.pos[1] / (2 ** octave))) + i
            if y > 0 and y < imgShape[0] - 1:
                for j in range(-radius, radius + 1):
                    x = int(round(keypoint.pos[0] / (2 ** octave)))
                    if x > 0 and x < imgShape[1] - 1:
                        dx = img[y, x + 1] - img[y, x - 1]
                        dy = img[y - 1, x] - img[y + 1, x]
                        magnitude = math.sqrt(dx * dx + dy * dy)
                        orientation = np.rad2deg(np.arctan2(dy, dx))
                        weight = np.exp(weight_factor * (i ** 2 + j ** 2))
                        histogram_index = int(round(orientation * num_bins / 360.))
                        raw_histogram[histogram_index % num_bins] += weight * magnitude
        #这个平滑公式根据OpenCV的公式            
        for n in range(num_bins):
            smooth_histogram[n] = (6 * raw_histogram[n] + 4 * (raw_histogram[n - 1] + raw_histogram[(n + 1) % num_bins]) + raw_histogram[n - 2] + raw_histogram[(n + 2) % num_bins]) / 16
        orientation_max = max(smooth_histogram)
        #这一步是寻找直方图中的局部最大值，就是在这个位置上的值比左邻的大也比右邻的大
        #因为以下的插值过程只需要大于peak_ratio * orientation_max的局部最大值，所以也可以有别的写法
        #np.where方法返回对应的index
        orientation_peaks = np.where(np.logical_and(smooth_histogram > np.roll(smooth_histogram, 1), smooth_histogram > np.roll(smooth_histogram, -1)))[0]
        #之后对于大概的方向进行插值精确处理，多项式插值后的多项式的极大值取代原本的值
        #并且要对每一个大于阈值的峰创建一个keypoint，就是copyu原来的keypoint，只不过方向不同
        for peak_index in orientation_peaks:
            peak_value = smooth_histogram[peak_index]
        if peak_value >= peak_ratio * orientation_max:
            # Quadratic peak interpolation
            # The interpolation update is given by equation (6.30) in https://ccrma.stanford.edu/~jos/sasp/Quadratic_Interpolation_Spectral_Peaks.html
            left_value = smooth_histogram[(peak_index - 1) % num_bins]
            right_value = smooth_histogram[(peak_index + 1) % num_bins]
            interpolated_peak_index = (peak_index + 0.5 * (left_value - right_value) / (left_value - 2 * peak_value + right_value)) % num_bins
            orientation = 360. - interpolated_peak_index * 360. / num_bins
            if math.isclose(orientation, 360.):
                orientation = 0
            keypoint.orientation = orientation
            keypoints_with_orientation.append(copy.deepcopy(keypoint))
    return keypoints_with_orientation
```

```python
#生成特征点描述符
def __generateDescriptors(keypoints, gaussian_image, threshold_after_normalization=0.2, window_width=4, num_bins=8, scale_multiplier=3, descriptor_max_value=0.2):
    descriptors = []

    for keypoint in keypoints:
        octave = keypoint.octave
        interval = keypoint.interval
        img = gaussian_image[octave, interval]
        imgShape = img.shape
        #这里要把特征点左边还原为在高斯模糊和降采样之后的图片里的坐标，为了保证能取到真实的附近的点
        posX = keypoint.pos[0] / (2 ** octave)
        posY = keypoint.pos[1] / (2 ** octave)
        angle = 360. - keypoint.orientation
        cos_angle = np.cos(np.deg2rad(angle))
        sin_angle = np.sin(np.deg2rad(angle))
        bins_per_degree = num_bins / 360
        weight_multiplier = -0.5 / ((0.5 * window_width) ** 2)
        sampling_data = []
        histogram_tensor = np.zeros((window_width + 2, window_width + 2, num_bins))
        # first two dimensions are increased by 2 to account for border effects
        
        #这里我保存的size是在关键点所属DoG图像的环境下的size，就是高斯函数的一个sigma，这里的采样遵守三sigma以外的点可以忽视的原则
        #所以这里的size不需要计算由图像的缩放（因为降采样）造成的大小变化
        seed_width = scale_multiplier * keypoint.size
        radius = int(round(scale_multiplier * keypoint.size * math.sqrt(2) * (window_width + 1) * 0.5))
        #要确保邻域的大小在图像范围内，否则没有像素点来取值
        radius = int(np.min([radius, math.sqrt(imgShape[0] ** 2 + imgShape[1] ** 2)]))
        
        for row in range(-radius, radius + 1):
            for col in range(-radius, radius + 1):
                row_rot = col * sin_angle + row * cos_angle
                col_rot = col * cos_angle - row * sin_angle
                # + 0.5 * window_width - 0.5全部是为了变换坐标系
                #在这里以一个子区域（种子点）的边长为整个坐标系的单位长度（单位长度变大）
                #坐标原点在左上角的种子点的中心
                #window_width描述的就是四个单位长度
                row_bin = (row_rot / seed_width) + 0.5 * window_width - 0.5
                col_bin = (col_rot / seed_width) + 0.5 * window_width - 0.5
                #检测坐标转换之后是否越界，坐标转换只是为了确定哪些点需要被采样
                if row_bin > -1 and row_bin < window_width and col_bin > -1 and col_bin < window_width:
                    #再给原来的坐标加上关键点的坐标作为偏移，得到真正应该采样的点，构成围绕关键点的邻域
                    #注意图像的坐标是y管row，x管col
                    window_row = int(round(posY + row))
                    window_col = int(round(posX + col))
                    #检测待采样的点是否超出图像边界，numpy的shaoe始终是0为row，1为col
                    if window_row > 0 and window_row < imgShape[0] - 1 and window_col > 0 and window_col < imgShape[1] - 1:
                        dx = img[window_row + 1, window_col] - img[window_row - 1, window_col]
                        dy = img[window_row, window_col + 1] - img[window_row, window_col - 1]
                        magnitude = math.sqrt(dx * dx + dy * dy)
                        orientation = np.rad2deg(np.arctan2(dy, dx)) % 360
                        weight = np.exp(weight_multiplier * ((row_rot / seed_width) ** 2 + (col_rot / seed_width) ** 2))
                        #减去angle是为了将角度变为以angle为0度的坐标系里，bins_per_degree代表一度代表多少bin，将度数转换为0-7这8个方向
                        orientation_bin = (orientation - angle) * bins_per_degree
                        sampling_data.append((row_bin, col_bin, magnitude * weight, orientation_bin))
                        
        for sample in sampling_data:
            row_bin_floor, col_bin_floor, orientation_bin_floor = np.floor([sample[0], sample[1], sample[3]]).astype(int)
            row_fraction = sample[0] - row_bin_floor  
            col_fraction = sample[1] - col_bin_floor
            orientation_fraction = sample[0] - orientation_bin_floor
            if orientation_bin_floor < 0:
                orientation_bin_floor += num_bins
            if orientation_bin_floor >= num_bins:
                orientation_bin_floor -= num_bins
            
            c1 = sample[2] * row_fraction
            c0 = sample[2] * (1 - row_fraction)
            c11 = c1 * col_fraction
            c10 = c1 * (1 - col_fraction)
            c01 = c0 * col_fraction
            c00 = c0 * (1 - col_fraction)
            c111 = c11 * orientation_fraction
            c110 = c11 * (1 - orientation_fraction)
            c101 = c10 * orientation_fraction
            c100 = c10 * (1 - orientation_fraction)
            c011 = c01 * orientation_fraction
            c010 = c01 * (1 - orientation_fraction)
            c001 = c00 * orientation_fraction
            c000 = c00 * (1 - orientation_fraction)
            #提前对bin的坐标系进行了转换
            histogram_tensor[row_bin_floor + 1, col_bin_floor + 1, orientation_bin_floor] += c000
            histogram_tensor[row_bin_floor + 1, col_bin_floor + 1, (orientation_bin_floor + 1) % num_bins] += c001
            histogram_tensor[row_bin_floor + 1, col_bin_floor + 2, orientation_bin_floor] += c010
            histogram_tensor[row_bin_floor + 1, col_bin_floor + 2, (orientation_bin_floor + 1) % num_bins] += c011
            histogram_tensor[row_bin_floor + 2, col_bin_floor + 1, orientation_bin_floor] += c100
            histogram_tensor[row_bin_floor + 2, col_bin_floor + 1, (orientation_bin_floor + 1) % num_bins] += c101
            histogram_tensor[row_bin_floor + 2, col_bin_floor + 2, orientation_bin_floor] += c110
            histogram_tensor[row_bin_floor + 2, col_bin_floor + 2, (orientation_bin_floor + 1) % num_bins] += c111
        descriptor_vector = histogram_tensor[1:-1, 1:-1, :].flatten()  # Remove histogram borders
        # Threshold and normalize descriptor_vector
        discriptor_sum = np.sum(descriptor_vector)
        descriptor_vector = descriptor_vector / math.sqrt(discriptor_sum)
        
        descriptor_vector[descriptor_vector > threshold_after_normalization] = threshold_after_normalization
        
        discriptor_sum = np.sum(descriptor_vector)
        #descriptor_vector = descriptor_vector / math.sqrt(discriptor_sum)
        # Multiply by 512, round, and saturate between 0 and 255 to convert from float32 to unsigned char (OpenCV convention)
        #descriptor_vector = round(512 * descriptor_vector)
        #descriptor_vector[descriptor_vector < 0] = 0
        #descriptor_vector[descriptor_vector > 255] = 255
        descriptors.append(descriptor_vector)
    return np.array(descriptors, dtype='float32')
```
